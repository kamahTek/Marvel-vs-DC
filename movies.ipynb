## Data Understanding

```python
# make a copy of the dataset for further analysis
entertainment_df = df.copy()
entertainment_df
```

```python
''' Identify if column ID has any important relevance in our dataset .
    It appears it is a repetition of the row index.
    Now we should drop it permanently
'''
entertainment_df['ID'].value_counts()
```

```python
# Drop column ID
entertainment_df = entertainment_df.drop(columns='ID')
entertainment_df
```

```python
# Understand construction of columns in the dataset
entertainment_df.info()
```

```python
# From the Observation Above:
'''The dataset has a total of 6 columns
   5 columns are of Dtype object -> Movie, Year, Genre, RunTime, Description
   1 column is of Dtype float -> IMDB_SCORE
'''
print(' ')
```

**NOTE**

Columns to convert from objects to int -> RunTime

Columns to convert from objects to Datetime -> Year

```python
# Clean RunTime column
# Strip additional spaces first before replacing data
entertainment_df['RunTime'] = entertainment_df['RunTime'].str.strip().str.replace(' min', '')
entertainment_df
```

```python
# convert RunTime column from object to integer type
entertainment_df['RunTime'] = entertainment_df['RunTime'].astype(int)
```

```python
# Cleaning column Year
entertainment_df['Year']
```

```python
# Remove parentheses
entertainment_df['Year'] = entertainment_df['Year'].str.replace(r'[()]', '', regex=True)  
entertainment_df['Year']
```

```python
# Extract individual years and year ranges
entertainment_df['Year'] = entertainment_df['Year'].str.extract(r'(\b\d{4}\b(?:–\d{4})?)')
entertainment_df['Year']
```

```python
# Separate year ranges into start and end years
entertainment_df['Start Year'] = entertainment_df['Year'].str.extract(r'(\b\d{4}\b)')
entertainment_df['End Year'] = entertainment_df['Year'].str.extract(r'–(\d{4})')
```

```python
# Avoid dropping of years with nan values
entertainment_df['End Year'] = entertainment_df['End Year'].replace(np.nan, 'not defined')
```

**Create a custom function for checking the mean, mode and median of dataset column**

```python
# CUSTOM FUNCTION
def m_m_m(col_name, stat_func):
    if stat_func == 'mean':
        col_stat = entertainment_df[col_name].mean()
    elif stat_func == 'median':
        col_stat = entertainment_df[col_name].median()
    elif stat_func == 'mode':
        col_stat = entertainment_df[col_name].mode()[0]  # Get the first mode value
    else:
        raise ValueError("Invalid stat_func. Use 'mean', 'median', or 'mode'.")
    
    return f"{stat_func.capitalize()}: {col_stat}"
```

```python
# check how many times runtime == 0. 
entertainment_df['RunTime'].value_counts()
```

```python
m_m_m('RunTime', 'mode')
```

```python
m_m_m('RunTime', 'mean')
```

```python
m_m_m('RunTime', 'median')
```

```python
# Replace the movies with 0 RunTime to be NaN
entertainment_df['RunTime'] = entertainment_df['RunTime'].replace(0, round(entertainment_df['RunTime'].mean(),0))
```
